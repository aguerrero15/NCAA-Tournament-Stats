```python
import glob

import numpy as np
import pandas as pd

path = './data'
csv_files = glob.glob(path + '/*.csv')

df_list = (pd.read_csv(file) for file in csv_files)

df = pd.concat(df_list, ignore_index=True)
```

```python
df['upset'] = np.where(df['Winner Seed'] > df['Loser Seed'], True, False)
df
```
